```python
# exciting_curves.py
import numpy as np
import matplotlib.pyplot as plt

# ---- Helpers ---------------------------------------------------------------
def prettify(ax, title=""):
    ax.set_aspect("equal")
    ax.set_title(title, pad=10)
    ax.set_xticks([]); ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)

def save_and_show(fig, name):
    fig.tight_layout()
    fig.savefig(f"{name}.png", dpi=300, bbox_inches="tight")
    plt.show()
    print(f"Saved: {name}.png")

# ---- Curves ----------------------------------------------------------------
def curve_lissajous(A=1, B=1, a=3, b=4, delta=np.pi/2, n=4000):
    t = np.linspace(0, 2*np.pi, n)
    x = A * np.sin(a*t + delta)
    y = B * np.sin(b*t)
    return x, y, f"Lissajous (a={a}, b={b}, δ=π/2)"

def curve_rose(k=5, n=4000):
    # r = cos(kθ). If k is integer: k petals (k odd), 2k petals (k even)
    t = np.linspace(0, 2*np.pi, n)
    r = np.cos(k*t)
    x = r * np.cos(t)
    y = r * np.sin(t)
    petals = k if k % 2 == 1 else 2*k
    return x, y, f"Rose curve (k={k}, petals={petals})"

def curve_hypotrochoid(R=5, r=3, d=5, n=6000):
    # Hypotrochoid: inside-rolling trochoid
    t = np.linspace(0, 2*np.pi*max(R, r), n)
    x = (R - r)*np.cos(t) + d*np.cos(((R - r)/r)*t)
    y = (R - r)*np.sin(t) - d*np.sin(((R - r)/r)*t)
    return x, y, f"Hypotrochoid (R={R}, r={r}, d={d})"

def curve_butterfly(n=6000):
    # "Butterfly" curve
    t = np.linspace(0, 12*np.pi, n)
    r = np.exp(np.sin(t)) - 2*np.cos(4*t) + (np.sin((2*t - np.pi)/24))**5
    x = r * np.cos(t)
    y = r * np.sin(t)
    return x, y, "Butterfly curve"

def curve_superformula(m=6, a=1, b=1, n1=0.3, n2=0.3, n3=0.3, n=6000):
    # Gielis superformula in 2D (beautiful variety)
    t = np.linspace(0, 2*np.pi, n)
    part1 = (1/a)*np.abs(np.cos(m*t/4))**n2
    part2 = (1/b)*np.abs(np.sin(m*t/4))**n3
    r = (part1 + part2)**(-1/n1)
    x = r * np.cos(t)
    y = r * np.sin(t)
    return x, y, f"Superformula (m={m}, n1={n1}, n2={n2}, n3={n3})"

# ---- Pick & Plot -----------------------------------------------------------
if __name__ == "__main__":
    # Choose one of: "lissajous", "rose", "hypotrochoid", "butterfly", "super"
    which = "super"   # <- change this to try different curves

    if which == "lissajous":
        x, y, title = curve_lissajous(A=1, B=1, a=5, b=4, delta=np.pi/2)
        name = "curve_lissajous"
    elif which == "rose":
        x, y, title = curve_rose(k=7)
        name = "curve_rose"
    elif which == "hypotrochoid":
        x, y, title = curve_hypotrochoid(R=7, r=3, d=4)
        name = "curve_hypotrochoid"
    elif which == "butterfly":
        x, y, title = curve_butterfly()
        name = "curve_butterfly"
    elif which == "super":
        # Try small tweaks for wildly different shapes:
        # e.g., m=7, n1=0.2, n2=1.7, n3=1.7 or m=3, n1=0.2, n2=0.8, n3=0.8
        x, y, title = curve_superformula(m=7, n1=0.2, n2=1.7, n3=1.7)
        name = "curve_superformula"
    else:
        raise ValueError("Unknown curve name.")

    fig, ax = plt.subplots(figsize=(6, 6))
    ax.plot(x, y, linewidth=1.2)
    prettify(ax, title)
    save_and_show(fig, name)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Helpers ---------------------------------------------------------------
def prettify(ax, title=""):
    ax.set_aspect("equal")
    ax.set_title(title, pad=8, fontsize=10)
    ax.set_xticks([]); ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)

# ---- Curves ----------------------------------------------------------------
def curve_lissajous(A=1, B=1, a=3, b=4, delta=np.pi/2, n=4000):
    t = np.linspace(0, 2*np.pi, n)
    x = A * np.sin(a*t + delta)
    y = B * np.sin(b*t)
    return x, y, f"Lissajous (a={a}, b={b}, δ=π/2)"

def curve_rose(k=5, n=4000):
    t = np.linspace(0, 2*np.pi, n)
    r = np.cos(k*t)
    x = r * np.cos(t)
    y = r * np.sin(t)
    petals = k if k % 2 == 1 else 2*k
    return x, y, f"Rose curve (k={k}, petals={petals})"

def curve_hypotrochoid(R=5, r=3, d=5, n=6000):
    t = np.linspace(0, 2*np.pi*max(R, r), n)
    x = (R - r)*np.cos(t) + d*np.cos(((R - r)/r)*t)
    y = (R - r)*np.sin(t) - d*np.sin(((R - r)/r)*t)
    return x, y, f"Hypotrochoid (R={R}, r={r}, d={d})"

def curve_butterfly(n=6000):
    t = np.linspace(0, 12*np.pi, n)
    r = np.exp(np.sin(t)) - 2*np.cos(4*t) + (np.sin((2*t - np.pi)/24))**5
    x = r * np.cos(t)
    y = r * np.sin(t)
    return x, y, "Butterfly curve"

def curve_superformula(m=6, a=1, b=1, n1=0.3, n2=0.3, n3=0.3, n=6000):
    t = np.linspace(0, 2*np.pi, n)
    part1 = (1/a)*np.abs(np.cos(m*t/4))**n2
    part2 = (1/b)*np.abs(np.sin(m*t/4))**n3
    r = (part1 + part2)**(-1/n1)
    x = r * np.cos(t)
    y = r * np.sin(t)
    return x, y, f"Superformula (m={m}, n1={n1}, n2={n2}, n3={n3})"

# ---- Main ------------------------------------------------------------------
if __name__ == "__main__":
    curves = [
        curve_lissajous(a=5, b=4),
        curve_rose(k=7),
        curve_hypotrochoid(R=7, r=3, d=4),
        curve_butterfly(),
        curve_superformula(m=7, n1=0.2, n2=1.7, n3=1.7)
    ]

    fig, axes = plt.subplots(2, 3, figsize=(12, 8))
    axes = axes.flatten()

    for i, (x, y, title) in enumerate(curves):
        axes[i].plot(x, y, linewidth=1.2)
        prettify(axes[i], title)
        # Save each curve as a separate PNG
        plt.figure(figsize=(5, 5))
        plt.plot(x, y)
        plt.gca().set_aspect("equal")
        plt.axis("off")
        plt.tight_layout()
        plt.savefig(f"{title.replace(' ', '_').replace('(', '').replace(')', '')}.png",
                    dpi=300, bbox_inches="tight")
        plt.close()

    # Hide extra subplot (because 2x3 grid = 6 but we have 5 curves)
    axes[-1].axis("off")

    fig.tight_layout()
    plt.show()
```
