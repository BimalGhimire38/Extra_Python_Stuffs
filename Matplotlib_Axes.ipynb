```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weekly Weather")
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend(loc="upper right", shadow=True) # best,
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b=a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y= 2*x
```

```python
x,y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)

plt.show()
```

```python
type(fig)
```

```python
fig = plt.figure()

# Large Axis

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

# Small Axes
axes2 = fig.add_axes([0.2,0.2,0.25,0.25])
axes2.plot(x,y)


plt.show()
```

```python
fig = plt.figure()

# Large Axis

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title("Power to 4")

# Small Axes
axes2 = fig.add_axes([0.2,0.2,0.25,0.25])

axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title("Zoomed in")


plt.show()
```
