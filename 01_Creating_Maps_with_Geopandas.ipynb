```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file("/content/sample_data/district_boundaries/district.shp")
```

```python
gdf.head()
```

# Basic Visualization

```python
gdf.plot()
```

```python
gdf.plot('Shape_Area')
```

```python
gdf.plot('Shape_Area', cmap='RdYlBu')
gdf.plot('Shape_Area', cmap='hsv',legend= True)
```

```python
gdf.plot('SUM_area', cmap='hsv',legend= True, vmin=1000, vmax=6000)
```

```python
gdf.plot('SUM_area', cmap='hsv',legend= True, vmin=1000, vmax=6000, edgecolor='black')
```

```python
gdf.plot('SUM_area', cmap='hsv',legend= True, vmin=1000, vmax=6000, edgecolor='black',
         legend_kwds = {
             'label':'Area (km2)'
         }
         )
```

```python
fig, ax = plt.subplots(1,1,figsize = (10,8))
gdf.plot('SUM_area', cmap='hsv', ax = ax, legend= True, vmin=1000, vmax=6000, edgecolor='black',
         legend_kwds = {
             'label':'Area (km2)'
         }
         )

ax.set_title("Nepal (Area)")
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

ax
```

```python
provinces = gdf['Province'].unique()
```

```python
provinces
```

```python
fig, ax = plt.subplots(2,4, figsize=(15,10))
axis = ax.flatten()

axis[7].remove()

for (ax, province) in zip(ax, provinces):
  print(ax, province)
```

```python
fig, ax = plt.subplots(2,4, figsize=(15,10))
axis = ax.flatten()

axis[7].remove()

for ax_item, province in zip(axis[:len(provinces)], provinces):
  df_province = gdf[gdf['Province']==province]

  df_province.plot('SUM_area', cmap='RdYlGn', ax=ax_item, edgecolor='black')

  ax_item.set_title(f"{province} Province of Nepal")

  ax_item.set_xticks([])
  ax_item.set_yticks([])
```
